# 1. Feature Engineering:


### ***Host_Tenure :***
* Creating '**Host_Tenure**' feature by calculating the Number of years from **host_since** to current date.
* it will help our model to measure the experience of host it means if Host has more Experience better he works.

```python

#function to calculate tenure
def calculate_tenure(date_is):

  days_difference = (datetime.now()-date_is).days
  tenure = round(days_difference /365 +0.5 , 1)
  return tenure

# Apply the function to create new column
data['Host_Tenure'] = data['host_since'].apply(calculate_tenure)

#display sample 5 records
data['Host_Tenure'][:5]
```

* we can see new feature is created **Host_tenure** it defines exeperience in years and after point its a months.
* for instance first record defines **12.6** it means 12 years and 6 months of experience.

# 4. Sentiment Analysis on Textual Data: 

* we are applying Advance natural language processing technique to the **desciption** text and extracting sentiment score.
* for this task we are using **TextBlob** library which is use for NLP tasks and it can analyze sentiments like **positive, negative or neutral** and it is easy to implement. this is perfect library for doing our task.
1. firstly are calculating **sentimaent score** of each **desciption** text using TextBlob and store it in new feature **sentiment_score**.
2. and then we are using sentiment_score for trainign **Linear regression model** to predict the **log_price**.
3.and then we will chech the **mean_suared_error**

```python
# Analyzing desciption text to preprocessing 
# randomely seeing 4 records in description to see is text well maintained or is in Lower case or upper case.
# we need to do preprocessing of text or not ??
n=1
for txt in data['description'].sample(4):
    print(n,txt)
    n+=1    
```

* as we can see texts is in well manner and not need to do any preprocessing text is ready to make sentiment score 

```python
#creating sentiment score feature.
from textblob import TextBlob

# Sentiment Analysis using TextBlob
data['sentiment_score'] = data['description'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Visualise the distribution of Sentiment score feature
plt.figure(figsize=(10,5))
sns.histplot(data['sentiment_score'],kde=True,color='seagreen')
plt.title("distribution of Sentimental_score ")
plt.show()
```

* if description is towards 1 it means **Possitive description**  and if towards -1 it means **negative description**
* and if description is 0 it means the **Neutral description**
* we can see the more desciptions in above plot is between **0 to 0.6** it means the more descriptions are **neutral or little pissitive**.

```python
# prediction of log price using sentiment score and performence mtrics.
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


#split data into training and testing sets 
x = data[['sentiment_score']]
y = data['log_price']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)

#Train linear Model
model = LinearRegression()
model.fit(x_train,y_train)

#make prediction
y_pred = model.predict(x_test)

#Model Evaluation
mean_se = mean_squared_error(y_test,y_pred)
print("Average squared difference between actual and predicted price is  : %1.2f"%mean_se)
```

* our Mean squared error is 0.51 which means **sentiment score** not perfectly but it will be a one parameter which help us to achive more goodness of our final model.

# 6. Categorical Data Encoding: 

#### Encoding : 
for better performance of oue model we need to conver Categorical features in numerical and here cames One Hote Encoding. this method converts categorical data into binary features where each unique value convertes in separete feature and theire data points are in 0 and 1.
1. as per task we are doing **One Hot Encoding** with **room_type**, **City** and **property_type** features.
2. other than that we have binary features like **cleaning_fee**,  **host_identity_verified**, **instant_bookable** so we are converting thos in 0 and 1 .
3. also we have **cancellation_policy** in which there are ordinal relationship between values it means ***flexible*** being less strict than ***moderate*** and ***moderate*** less strict than ***strict*** like wise so we are doing **ordinal Encoding** on cancellation polity.
* 'flexible' -> 0 
'moderate' -> 1
 'strict' -> 2
 'super_strict_30' -> 3
'super_strict_60' -> 4

#### 1. One Hot Encoding with **room_type**, **City** and **property_type**

```python
# One Hote Encoding using get dummies
# 1) room_type
# getting dummies and set new column name on values. 
roomtype_encoded = pd.get_dummies(data['room_type'],prefix='room_type',dtype=int)
 # concatenating the encoded feature with main data.
data = pd.concat([data,roomtype_encoded],axis=1)


# 2) city
# getting dummies and set new column name on values. 
city_encoded = pd.get_dummies(data['city'],prefix='city',dtype=int)
#concatenating the encoded feature 
data = pd.concat([data,city_encoded],axis=1)

# 3) property_type
# getting dummies and set new column name on values. 
property_encoded = pd.get_dummies(data['property_type'],prefix='property_type',dtype=int)
#concatenating the encoded feature 
data = pd.concat([data,property_encoded],axis=1)

#drop room_type feature
del data['room_type']
#drop city feature
del data['city']
# drop property_type feature
del data['property_type']

#display data 
data.sample(2)
```

#### 2. Binary features encoding  **cleaning_fee**, **host_identity_verified**, **instant_bookable**

```python
#1) cleaning fee 
# Datatype of cleaning fee is bool (binary) so we can directly convert into (0 & 1) using astype()
data['cleaning_fee'] = data['cleaning_fee'].astype(int)
print(data['cleaning_fee'].value_counts())


# 2) instant_bookable 
# we are replacing 't' with 1and 'f' as 0 
data['instant_bookable'] = data['instant_bookable'].replace('t',1)
data['instant_bookable'] = data['instant_bookable'].replace('f',0)
print(data['instant_bookable'].value_counts())


# 3) host_identity_verified 
# we are replacing 't' with 1and 'f' as 0 
data['host_identity_verified'] = data['host_identity_verified'].replace('t',1)
data['host_identity_verified'] = data['host_identity_verified'].replace('f',0)
print(data['host_identity_verified'].value_counts())
```

### 2.  RandomForest - Non linear - ensamble technique.

* Random Forest is a ensemble technique/ bagging technique based on decision tree it combines the prediction of multiple individuals trees to increas accuracy. it simply create multiple subset of training data using process bootstraping sampleng method and each subset is used for train decisionn tree. in regression it takes a average/ mean of all dicison trees.
* why random forest **because it robust to overfitting and work well with unseen data/ validation data(y_test).
* in our contex **Dataset is so huge and we have so many features so random forest work well with high dimensional data**.


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Initialize the random forest regressor
# defaul criterion is use for measur the quality of a split.default criterial is "Mean squared error"
# which means it measure MSE of target variable in each split. 
random_f = RandomForestRegressor( n_estimators=100,random_state=42,max_depth=2)

# Training the random forest
RF_model = random_f.fit(x_train,y_train)

#make prediction on tesing data
RF_ypred = RF_model.predict(x_test)

# calculating mean squared error
rfmse =mean_squared_error(y_test,RF_ypred)
print("mean squared error : %.4f "%rfmse)
# root mean squared error 
print("root mean squared error : %.4f "%np.sqrt(rfmse))
# r squared 
print("r2_score : %.4f "%r2_score(y_test,RF_ypred))
```

* we are fit the model of RAndom forest default quality measure parameter is ()(criterion) is **mean squared error**

* random stae 42 means it iteratively pick the sample after 42.
* in the depth of decision tree we got **Overfitting** so we did **pre prunning** which means selecting the  **Max_depth** means stopping the tree in 3th depth of tree to avoid overfitting.

# 8. Model Optimization and Validation: 

* **GridSearchCV** : is a method provided by scikit-learn for performing hyperparameter tuning through an exhaustive search over a specified parameter grid. It systematically evaluates the model performance for each combination of hyperparameters using **cross-validation** and selects the best combination based on a scoring metric.
* **Cross-Validation** : GridSearchCV splits the training data into multiple folds (default is 5-fold) and trains the model on different combinations of hyperparameters using each fold for validation. This allows for an unbiased evaluation of the model's performance

1. Linear regression hyperparameter tunung :
* It seems that sklearn.linear_model.LinearRegression does not have hyperparameters that can be tuned. So we only work on remaining two models. 

2. Random Forest : hyperparameter tuning.

```python
from sklearn.model_selection import GridSearchCV


# Define hyperparameters to tune
param_grid = {
    'n_estimators': [100, 200, 300],            # Number of trees in the forest
    'max_depth': [None, 10, 20],                 # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],             # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4]                # Minimum number of samples required at each leaf node
}

# Perform Grid Search Cross Validation to find the best hyperparameters
grid_search = GridSearchCV(estimator=RF_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)

# Print the best hyperparameters found
print("Best hyperparameters:", grid_search.best_params_)

# Evaluate the model with the best hyperparameters on the test set
best_rf_model = grid_search.best_estimator_
test_score = best_rf_model.score(x_test, y_test)
print("Test set R^2 score with best hyperparameters:", test_score)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

esr = GradientBoostingRegressor()

parameters = {
    'learning_rate' : [0.001,0.01, 0.2],
    'n_estimators' : [50,150,180],
    'max_depth' : [2,4,6]
    
}

# Perform Grid Search with cross-validation
grid_search  = GridSearchCV(estimator=esr, param_grid=parameters, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x_train,y_train)


# Get the best parameters
best_params = grid_search.best_params_


# Train the model with the best parameters
best_model = GradientBoostingRegressor(**best_params)
best_model.fit(x_train, y_train)



# Evaluate the model on the test set
y_pred = best_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
